# Importing Libraries and Database connection:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
```

```python
# Path to the SQLite database and  Connect to the database
connection = sqlite3.connect("accident_data_v1.0.0_2023.db")
cursor = connection.cursor()
```

```python
print(cursor)
```

# Data Extraction for the year 2020:

```python
# Define the queries
queries = {
    'accidents_2020': "SELECT * FROM accident WHERE accident_year = 2020;",
    'casualties_2020': "SELECT * FROM casualty WHERE accident_year = 2020;",
    'vehicles_2020': "SELECT * FROM vehicle WHERE accident_year = 2020;"
}

# Execute the queries and store the results in DataFrames
accidents_2020, casualties_2020, vehicles_2020 = [pd.read_sql_query(query, connection) for query in queries.values()]

# Display the results
print("Accidents data for the year 2020:")
display(accidents_2020)
```

```python
print("\nCasualties data for the year 2020:")
display(casualties_2020)
```

```python
print("\nVehicles data for the year 2020:")
display(vehicles_2020)
```

# Task 1:

# Hour of the day and day of week with maximum accident:

```python
# 'date' column is in 'day/month/year' format 
accidents_2020['date'] = pd.to_datetime(accidents_2020['date'], format='%d/%m/%Y', dayfirst=True)

# Convert the 'time' field to datetime and extract hour
accidents_2020['hour_of_day'] = pd.to_datetime(accidents_2020['time'], format='%H:%M').dt.hour

# Extract day of the week from 'date'
accidents_2020['day_of_week'] = accidents_2020['date'].dt.day_name()

# Group by hour and day of the week, and count the number of accidents
accident_counts_hour = accidents_2020.groupby(['hour_of_day']).size()
accident_counts_week = accidents_2020.groupby(['day_of_week']).size()

# Maximum accidents by hour
max_hour = accident_counts_hour.idxmax()
max_accidents_hour = accident_counts_hour[max_hour]
print(f"Hour with maximum accidents: {max_hour}, Count: {max_accidents_hour}")

# Maximum accidents by day of week
max_day = accident_counts_week.idxmax()
max_accidents_week = accident_counts_week[max_day]
print(f"Day with maximum accidents: {max_day}, Count: {max_accidents_week}")

# Visualization
plt.figure(figsize=(14, 6))

# Accidents by Hour of the Day
plt.subplot(1, 2, 1)
sns.barplot(x=accident_counts_hour.index, y=accident_counts_hour.values)
plt.title('Accidents by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')

# Accidents by Day of the Week
plt.subplot(1, 2, 2)
sns.barplot(x=accident_counts_week.index, y=accident_counts_week.values, 
            order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Accidents by Day of the Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Accidents')

plt.tight_layout()
plt.show()
```

# Multi-Factor Analysis (Day of week)

 # CASE 1: Top 10 accidents on the basis of local_authority_ons_district and weather_conditions.

```python
# Group by local authority, weather conditions, and day of the week, and count the number of accidents
accident_counts_local_weather_day = accidents_2020.groupby(
    ['local_authority_ons_district', 'weather_conditions', 'day_of_week'],
    observed=False  
).size()

# Find the top 10 combinations with the highest total number of accidents. Firstly, getting the total count of accidents per combination of local authority and weather conditions
total_accident_counts = accidents_2020.groupby(
    ['local_authority_ons_district', 'weather_conditions'],
    observed=False 
).size()

# Get the top 10 combinations
top_10_accidents = total_accident_counts.nlargest(10).reset_index(name='Accident Count')

# For each of the top 10 combinations, find the day of the week with the highest accident count
def find_max_day_of_week(row):
    district = row['local_authority_ons_district']
    weather = row['weather_conditions']
    
    # Filter data for the current district and weather condition
    filtered_data = accidents_2020[(accidents_2020['local_authority_ons_district'] == district) & 
                                   (accidents_2020['weather_conditions'] == weather)]
    
    # Group by day of the week and count the number of accidents
    accident_counts_by_day = filtered_data.groupby('day_of_week').size()
    
    # Find the day of the week with the maximum number of accidents
    max_day = accident_counts_by_day.idxmax()
    max_accidents_day = accident_counts_by_day.max()
    
    return pd.Series([max_day, max_accidents_day], index=['day_of_week', 'max_day_accidents'])

# Apply the function to each row in the top_10_accidents DataFrame
top_10_accidents[['day_of_week', 'max_day_accidents']] = top_10_accidents.apply(find_max_day_of_week, axis=1)

# Drop the extra column used for calculation
top_10_accidents = top_10_accidents.drop(columns='max_day_accidents')

# Print the resulting DataFrame
print(top_10_accidents)
```

# Distributions of accidents by Hour For motorbikes

```python
motorbike_types = {3: "Motorcycle 125cc and under", 4: "Motorcycle over 125cc and up to 500cc", 5: "Motorcycle over 500cc"}

# SQL query 
motorbikes_accident_2020 = pd.read_sql_query("""
    SELECT * FROM accident 
    LEFT JOIN vehicle ON accident.accident_index = vehicle.accident_index 
    WHERE accident.accident_year = 2020 AND vehicle.vehicle_type IN (3, 4, 5)
""", connection)

# Converting the date and time conversion
motorbikes_accident_2020['date'] = pd.to_datetime(motorbikes_accident_2020['date'], dayfirst=True)
motorbikes_accident_2020['time'] = pd.to_datetime(motorbikes_accident_2020['time'], format='%H:%M').dt.time

```

```python
# convert vehicle type to descriptive names:
motorbikes_accident_2020['vehicle_types'] = motorbikes_accident_2020['vehicle_type'].map(motorbike_types) 
```

```python
# Convertting the time conversion
motorbikes_accident_2020['time'] = pd.to_datetime(motorbikes_accident_2020['time'], format='%H:%M:%S')

# Extracting the hour from the time column
motorbikes_accident_2020['hour'] = motorbikes_accident_2020['time'].dt.hour

# Mapping vehicle types to their descriptions
vehicle_type_mapping = {
    3: "Motorcycle 125cc and under",
    4: "Motorcycle over 125cc and up to 500cc",
    5: "Motorcycle over 500cc"
}
motorbikes_accident_2020['vehicle_type_desc'] = motorbikes_accident_2020['vehicle_type'].map(vehicle_type_mapping)

# Plotting the distribution of accidents by hour of the day
plt.figure(figsize=(10, 6))
sns.countplot(x='hour', hue='vehicle_type_desc', data=motorbikes_accident_2020, palette='Set1')
plt.title('Accident Count by Hour of Day for Motorbikes')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')
plt.legend(title='Vehicle Type')
plt.show()
```

# Plotting which weather condition is responsible for max.number of accidents (in an hour) for motorcycles. 

```python
# Function to find the weather condition with maximum accidents for a specific vehicle type and hour
def find_max_weather_condition(data, vehicle_type, hour):
    filtered_data = data[(data['vehicle_type_desc'] == vehicle_type) & (data['hour'] == hour)]
    accident_counts_by_weather = filtered_data.groupby('weather_conditions').size().reset_index(name='Accident Count')
    max_accidents_weather = accident_counts_by_weather.loc[accident_counts_by_weather['Accident Count'].idxmax()]
    return max_accidents_weather, accident_counts_by_weather

# Custom legend labels
weather_condition_map = {
    1: "Fine without high winds",
    2: "Raining without high winds",
    3: "Snowing without high winds",
    4: "Fine with high winds",
    5: "Raining with high winds",
    6: "Snowing with high winds",
    7: "Fog or mist - if hazard",
    8: "Other",
    9: "Unknown"
}

# Function to plot accidents by weather condition
def plot_accidents_by_weather(accident_counts_by_weather, title, vehicle_type, hour):
    plt.figure(figsize=(12, 6))
    sns.barplot(x='weather_conditions', y='Accident Count', data=accident_counts_by_weather, hue=None)
    plt.title(title)
    plt.xlabel('Weather Conditions')
    plt.ylabel('Number of Accidents')
    plt.xticks(ticks=range(len(weather_condition_map)), 
               labels=[f"{key}" for key in weather_condition_map.keys()],
               rotation=45)
    
    # Add custom legend text manually
    legend_x = 0.75
    legend_y = 0.95
    legend_title = 'Weather Conditions'
    legend_text = "\n".join([f"{key}: {value}" for key, value in weather_condition_map.items()])
    
    # Plot custom legend with heading
    plt.text(legend_x, legend_y, f"{legend_title}\n{legend_text}", transform=plt.gca().transAxes, fontsize=10, va='top', ha='left', bbox=dict(boxstyle='round', facecolor='white', edgecolor='black'))

    plt.tight_layout()
    plt.show()

# 1. Analyze and plot for "Motorcycle 125cc and under" at the 18th hour
max_accidents_weather_125cc, accident_counts_by_weather_125cc = find_max_weather_condition(motorbikes_accident_2020, "Motorcycle 125cc and under", 18)
print("Weather condition with maximum accidents at 18th hour for Motorcycle 125cc and Under:")
print(max_accidents_weather_125cc)

plot_accidents_by_weather(accident_counts_by_weather_125cc, 'Accidents at 18th Hour of the Day for Motorcycle 125cc and Under by Weather Condition', "Motorcycle 125cc and under", 18)

# 2. Analyze and plot for "Motorcycle over 500cc" at the 17th hour
max_accidents_weather_500cc, accident_counts_by_weather_500cc = find_max_weather_condition(motorbikes_accident_2020, "Motorcycle over 500cc", 17)
print("Weather condition with maximum accidents at 17th hour for Motorcycle over 500cc:")
print(max_accidents_weather_500cc)

plot_accidents_by_weather(accident_counts_by_weather_500cc, 'Accidents at 17th Hour of the Day for Motorcycle over 500cc by Weather Condition', "Motorcycle over 500cc", 17)

# 3. Analyze and plot for "Motorcycle over 125cc and up to 500cc" at the 17th hour
max_accidents_weather_125_500cc, accident_counts_by_weather_125_500cc = find_max_weather_condition(motorbikes_accident_2020, "Motorcycle over 125cc and up to 500cc", 17)
print("Weather condition with maximum accidents at 17th hour for Motorcycle over 125cc and up to 500cc:")
print(max_accidents_weather_125_500cc)

plot_accidents_by_weather(accident_counts_by_weather_125_500cc, 'Accidents at 17th Hour of the Day for Motorcycle over 125cc and up to 500cc by Weather Condition', "Motorcycle over 125cc and up to 500cc", 17)
```

# Task 4:  Apriori algorithm

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Select relevant columns from accidents_2020
df = accidents_2020[['accident_severity', 'road_type', 'light_conditions', 'weather_conditions', 'time', 'date']].copy()

# Convert 'time' and 'date' columns to datetime with the correct format
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y')  # Keep as datetime object
df['time'] = pd.to_datetime(df['time'], format='%H:%M')  # Keep as datetime object

# Create 'hour' and 'day_of_week' columns
df['hour'] = df['time'].dt.hour.astype(str)
df['day_of_week'] = df['date'].dt.day_name()

# Select relevant columns for analysis
columns_of_interest = ['accident_severity', 'road_type', 'light_conditions', 'weather_conditions', 'hour', 'day_of_week']
df_col_int = df[columns_of_interest]

# One-hot encode the variables
df_encoded = pd.get_dummies(df_col_int, columns=columns_of_interest)

# Display the encoded data
display(df_encoded)
```

```python
# Apply the Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(df_encoded, min_support=0.01, use_colnames=True)

# Sort the frequent itemsets by support in descending order
frequent_itemsets_sorted = frequent_itemsets.sort_values(by='support', ascending=False).reset_index(drop=True)

# Display the sorted frequent itemsets
display(frequent_itemsets_sorted)
```

```python
# Sort the frequent itemsets by support and Plot the top 10 frequent itemsets
top_frequent_itemsets = frequent_itemsets.nlargest(10, 'support')
top_frequent_itemsets['itemsets'] = top_frequent_itemsets['itemsets'].apply(lambda x: ', '.join(list(x)))
plt.figure(figsize=(12, 6))
sns.barplot(x='support', y='itemsets', data=top_frequent_itemsets, hue='itemsets', legend=False, palette='viridis')
plt.title('Top 10 Frequent Itemsets')
plt.xlabel('Support')
plt.ylabel('Itemsets')
plt.show()
```

```python
# Generate the association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.2)

# Display the association rules for each accident severity
for col in df['accident_severity'].unique():
    # Filter rules related to accident severity
    severity_rules = rules[rules['consequents'].apply(lambda x: f'accident_severity_{col}' in str(x))]
    # Sort rules by confidence
    severity_rules = severity_rules.sort_values(by='lift', ascending=False)
    
    # Display the association rules
    display(severity_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

# Combining all above together, then applying TIME SERIES MODEL.

```python
# Execute the queries and store the results in DataFrame
accidents_df = pd.read_sql_query("SELECT * FROM accident", connection)

# Convert the 'date' column to datetime
accidents_df['date'] = pd.to_datetime(accidents_df['date'], format='%d/%m/%Y')

# Filter the DataFrame for selected police forces and date range
selected_police_forces = [1, 20, 46]
filtered_df = accidents_df[
    (accidents_df['police_force'].isin(selected_police_forces)) &
    (accidents_df['date'].between('2017-01-01', '2019-12-31'))
]

# Calculate weekly accident counts
weekly_counts = filtered_df.groupby(['police_force', pd.Grouper(key='date', freq='W-MON')]).size().reset_index(name='count')
weekly_counts_pivot = weekly_counts.pivot(index='date', columns='police_force', values='count').fillna(0)

# Convert PeriodIndex to DatetimeIndex with explicit frequency
weekly_counts_pivot.index = pd.to_datetime(weekly_counts_pivot.index, format='%Y-%m-%d')
weekly_counts_pivot.index.freq = 'W-MON'  
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error



# Define a function to fit ARIMA model and forecast
def train_arima_forecast(series, forecast_periods=52):
    model = ARIMA(series, order=(5, 0, 0))
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=forecast_periods)
    return forecast, model_fit

# Prepare for forecasting
forecast_periods = 52  # Number of weeks to forecast

# Loop through each selected police force and forecast
forecasts = {}
mse_scores = {}
for police_id in selected_police_forces:
    series = weekly_counts_pivot[police_id]
    forecast, model_fit = train_arima_forecast(series, forecast_periods)
    forecasts[police_id] = forecast

    # Calculate MSE
    actual = series[-forecast_periods:]
    mse = mean_squared_error(actual, forecast)
    mse_scores[police_id] = mse

# Display forecasts, historical data, and MSE scores
for police_id, forecast in forecasts.items():
    plt.figure(figsize=(12, 6))

    # Historical data
    plt.plot(weekly_counts_pivot.index, weekly_counts_pivot[police_id], label='Historical Data')

    # Future dates for forecast
    future_index = pd.date_range(start=weekly_counts_pivot.index[-1] + pd.DateOffset(weeks=1), periods=forecast_periods, freq='W-MON')

    # Forecast data
    plt.plot(future_index, forecast, label='Forecast', color='red')

    plt.title(f'Weekly Accident Counts Forecast for Police Force {police_id}')
    plt.xlabel('Date')
    plt.ylabel('Accidents')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Mean Squared Error for Police Force {police_id}: {mse_scores[police_id]:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
import warnings

# Suppress all warnings
warnings.filterwarnings("ignore")



# Define a function to fit SARIMA model and forecast
def train_sarima_forecast(series, forecast_periods=52):
    model = SARIMAX(series,
                    order=(5, 0, 0),           # ARIMA order (p, d, q)
                    seasonal_order=(1, 1, 1, 52))  # Seasonal order (P, D, Q, S) for weekly seasonality
    model_fit = model.fit(disp=False)
    forecast = model_fit.forecast(steps=forecast_periods)
    return forecast, model_fit

# Prepare for forecasting
forecast_periods = 52  # Number of weeks to forecast

# Loop through each selected police force and forecast
forecasts = {}
mse_scores = {}
for police_id in selected_police_forces:
    series = weekly_counts_pivot[police_id]
    forecast, model_fit = train_sarima_forecast(series, forecast_periods)
    forecasts[police_id] = forecast

    # Calculate MSE
    actual = series[-forecast_periods:]
    mse = mean_squared_error(actual, forecast)
    mse_scores[police_id] = mse

# Display forecasts, historical data, and MSE scores
for police_id, forecast in forecasts.items():
    plt.figure(figsize=(12, 6))

    # Historical data
    plt.plot(weekly_counts_pivot.index, weekly_counts_pivot[police_id], label='Historical Data')

    # Future dates for forecast
    future_index = pd.date_range(start=weekly_counts_pivot.index[-1] + pd.DateOffset(weeks=1), periods=forecast_periods, freq='W-MON')

    # Forecast data
    plt.plot(future_index, forecast, label='Forecast', color='red')

    plt.title(f'Weekly Accident Counts Forecast for Police Force {police_id}')
    plt.xlabel('Date')
    plt.ylabel('Accidents')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Mean Squared Error for Police Force {police_id}: {mse_scores[police_id]:.2f}")
```

```python
# Define the SQL query to find the top 3 LSOAs with the most accidents in Kingston upon Hull
query = """
SELECT 
    lsoa.lsoa01cd, 
    COUNT(*) AS accident_count
FROM 
    accident
LEFT JOIN 
    lsoa ON accident.lsoa_of_accident_location = lsoa.lsoa01cd
WHERE 
    lsoa.lsoa01nm LIKE '%Kingston upon Hull%'
    AND strftime('%Y-%m-%d', substr(accident.date, 7, 4) || '-' || substr(accident.date, 4, 2) || '-' || substr(accident.date, 1, 2)) BETWEEN '2020-01-01' AND '2020-03-31'
GROUP BY 
    lsoa.lsoa01cd
ORDER BY 
    accident_count DESC
LIMIT 3
"""

# Execute the query and load the results into a DataFrame
top_lsoas = pd.read_sql_query(query, connection)

# Display the top 3 LSOAs
print("Top 3 LSOAs with the highest number of accidents:")
print(top_lsoas)
```

```python
# Filter the accident data for the top 3 LSOAs and the date range
filtered_accidents = pd.read_sql_query(f"""
SELECT 
    strftime('%Y-%m-%d', substr(date, 7, 4) || '-' || substr(date, 4, 2) || '-' || substr(date, 1, 2)) AS date, 
    COUNT(*) AS accident_count
FROM 
    accident
WHERE 
    lsoa_of_accident_location IN ({','.join(['?']*len(top_lsoas['lsoa01cd']))})
    AND strftime('%Y-%m-%d', substr(date, 7, 4) || '-' || substr(date, 4, 2) || '-' || substr(date, 1, 2)) BETWEEN '2020-01-01' AND '2020-03-31'
GROUP BY 
    date
ORDER BY 
    date
""", connection, params=top_lsoas['lsoa01cd'].tolist())

# Convert the 'date' column to datetime format
filtered_accidents['date'] = pd.to_datetime(filtered_accidents['date'])

# Define the start and end dates
start_date = '2020-01-01'
end_date = '2020-03-31'

# Create a date range with all dates between start_date and end_date
date_range = pd.date_range(start=start_date, end=end_date, freq='D')

# Create a DataFrame from the date range
dates_df = pd.DataFrame(date_range, columns=['date'])

# Merge the dates_df with filtered_accidents to include all dates
# This will fill missing dates in filtered_accidents with NaN in 'accident_count'
full_data = pd.merge(dates_df, filtered_accidents, on='date', how='left')

# Fill NaN values in 'accident_count' with 0
full_data['accident_count'] = full_data['accident_count'].fillna(0)

# Display the resulting DataFrame
print("DataFrame with all dates between 2020-01-01 and 2020-03-31 and corresponding accident counts:")
print(full_data)
```

```python


# SQL query to get accident data for the top 3 LSOAs from January to June
query = """
SELECT 
    strftime('%Y-%m-%d', substr(date, 7, 4) || '-' || substr(date, 4, 2) || '-' || substr(date, 1, 2)) AS date, 
    COUNT(*) AS accident_count
FROM 
    accident
WHERE 
    lsoa_of_accident_location IN ({})
    AND strftime('%Y-%m-%d', substr(date, 7, 4) || '-' || substr(date, 4, 2) || '-' || substr(date, 1, 2)) BETWEEN '2020-01-01' AND '2020-06-30'
GROUP BY 
    date
ORDER BY 
    date
""".format(','.join(['?']*len(top_lsoas['lsoa01cd'])))

# Execute the query and load the results into a DataFrame
six_months_data = pd.read_sql_query(query, connection, params=top_lsoas['lsoa01cd'].tolist())

# Convert the 'date' column to datetime format
six_months_data['date'] = pd.to_datetime(six_months_data['date'])

# Ensure the 'date' column is in datetime format and set it as the index with a frequency
six_months_data.set_index('date', inplace=True)
six_months_data = six_months_data.asfreq('D')  # Set frequency to daily

# Fill NaN values in 'accident_count' with 0
six_months_data['accident_count'] = six_months_data['accident_count'].fillna(0)

# Display the six months aggregated data
print("Aggregated accident counts from January to June:")
print(six_months_data)

```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import pandas as pd
import matplotlib.pyplot as plt

# Define a function to fit the ARIMA model and forecast
def forecast_arima(series, steps=31):
    model = ARIMA(series, order=(1, 0, 1))  # Order can be tuned based on model diagnostics
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=steps)
    return forecast, model_fit

# Assuming six_months_data is already defined and filled with accident counts from Jan-Jun
train_data = six_months_data['accident_count']  # Use all available data for training

# Fit the ARIMA model and get the model object
forecasted_values, model_fit = forecast_arima(train_data, steps=31)

# Create a date range for the forecasted values (July)
forecast_dates = pd.date_range(start='2020-07-01', periods=31, freq='D')

# Combine the forecasted values with the dates
forecast_df = pd.DataFrame({'date': forecast_dates, 'forecasted_accident_count': forecasted_values})

# Calculate the MSE for the training data
train_mse = mean_squared_error(train_data, model_fit.fittedvalues)
print(f"Mean Squared Error (MSE) for training data: {train_mse:.2f}")

# Since we're forecasting future data, we don't have actual values for July
# We can't calculate MSE without actual data, but we'll visualize the forecast
plt.figure(figsize=(12, 6))
plt.plot(six_months_data.index, six_months_data['accident_count'], label='Historical Accident Counts (Jan-Jun)')
plt.plot(forecast_df['date'], forecast_df['forecasted_accident_count'], label='Forecasted Accident Counts (July)', linestyle='--')
plt.title('Historical and Forecasted Daily Accident Counts (Top 3 LSOAs Combined)')
plt.xlabel('Date')
plt.ylabel('Accident Counts')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
import pandas as pd
import matplotlib.pyplot as plt

# Define a function to fit the SARIMA model and forecast
def forecast_sarima(series, steps=31):
    model = SARIMAX(series, 
                    order=(1, 0, 1),           # ARIMA order (p, d, q)
                    seasonal_order=(1, 1, 1, 30))  # Seasonal order (P, D, Q, S), assuming weekly seasonality
    model_fit = model.fit(disp=False)
    forecast = model_fit.forecast(steps=steps)
    return forecast, model_fit

# Assuming six_months_data is already defined and filled with accident counts from Jan-Jun
train_data = six_months_data['accident_count']  # Use all available data for training

# Fit the SARIMA model and get the model object
forecasted_values, model_fit = forecast_sarima(train_data, steps=31)

# Create a date range for the forecasted values (July)
forecast_dates = pd.date_range(start='2020-07-01', periods=31, freq='D')

# Combine the forecasted values with the dates
forecast_df = pd.DataFrame({'date': forecast_dates, 'forecasted_accident_count': forecasted_values})

# Calculate the MSE for the training data
train_mse = mean_squared_error(train_data, model_fit.fittedvalues)
print(f"Mean Squared Error (MSE) for training data: {train_mse:.2f}")

# Plot historical data and forecasted values
plt.figure(figsize=(12, 6))
plt.plot(six_months_data.index, six_months_data['accident_count'], label='Historical Accident Counts (Jan-Jun)')
plt.plot(forecast_df['date'], forecast_df['forecasted_accident_count'], label='Forecasted Accident Counts (July)', linestyle='--')
plt.title('Historical and Forecasted Daily Accident Counts (Top 3 LSOAs Combined)')
plt.xlabel('Date')
plt.ylabel('Accident Counts')
plt.legend()
plt.show()
```

```python
import networkx as nx

# Load the graph data from the file
file_path = 'facebook_combined.txt'
G = nx.read_edgelist(file_path, create_using=nx.Graph(), nodetype=int)

# Basic Network Characteristics
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()
density = nx.density(G)
average_degree = sum(dict(G.degree()).values()) / num_nodes

print(f"Number of Nodes: {num_nodes}")
print(f"Number of Edges: {num_edges}")
print(f"Network Density: {density}")
print(f"Average Degree: {average_degree}")
```

```python
# Simplified visualization of the network by plotting a small subgraph

subgraph = G.subgraph(list(G.nodes)[:100])

plt.figure(figsize=(12, 12))
pos = nx.spring_layout(subgraph, seed=42)
nx.draw(subgraph, pos, node_size=50, with_labels=False, edge_color='gray', alpha=0.7)
plt.title("Social Network Subgraph Visualization (100 Nodes)")
plt.show()
```

```python
from community import community_louvain
print(f"Graph loaded: {G.number_of_nodes()} nodes and {G.number_of_edges()} edges\n")

# Label Propagation Algorithm (LPA)
print("Running Label Propagation Algorithm (LPA)...")
lpa_communities = list(nx.community.label_propagation_communities(G))
lpa_num_communities = len(lpa_communities)
lpa_community_sizes = [len(c) for c in lpa_communities]
print(f"LPA detected {lpa_num_communities} communities\n")

# Print LPA community details
print("LPA Community Details:")
print(f"Number of communities: {lpa_num_communities}")
for i, community in enumerate(sorted(lpa_communities, key=len, reverse=True)):
    print(f"Community {i}: {len(community)} nodes")

# Louvain Method
print("\nRunning Louvain Method...")
louvain_communities = community_louvain.best_partition(G)
louvain_community_groups = {}
for node, community in louvain_communities.items():
    louvain_community_groups.setdefault(community, []).append(node)
louvain_num_communities = len(louvain_community_groups)
louvain_community_sizes = [len(c) for c in louvain_community_groups.values()]
print(f"Louvain Method detected {louvain_num_communities} communities\n")

# Print Louvain community details
print("Louvain Community Details:")
print(f"Number of communities: {louvain_num_communities}")
for i, (community, nodes) in enumerate(sorted(louvain_community_groups.items(), key=lambda item: len(item[1]), reverse=True)):
    print(f"Community {i}: {len(nodes)} nodes")

# Plot the community size distributions
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.bar(range(lpa_num_communities), sorted(lpa_community_sizes, reverse=True), color='blue', alpha=0.7)
plt.title('Label Propagation Algorithm (LPA) Community Sizes')
plt.xlabel('Community')
plt.ylabel('Size')

plt.subplot(1, 2, 2)
plt.bar(range(louvain_num_communities), sorted(louvain_community_sizes, reverse=True), color='green', alpha=0.7)
plt.title('Louvain Community Sizes')
plt.xlabel('Community')
plt.ylabel('Size')

plt.tight_layout()
plt.show()
```
