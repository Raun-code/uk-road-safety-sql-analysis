```python
# Police Force 1: ID 16 (Humberside), Police Force 2: ID 13 (West Yorkshire), Police Force 3: ID 1 (Metropolitan Police)
selected_police_forces = [16, 13, 1]
filtered_df = accidents_df[(accidents_df['police_force'].isin(selected_police_forces)) & (accidents_df['accident_year'].between(2017, 2019))].copy()

# Convert the date to datetime format and set it as the index
filtered_df['date'] = pd.to_datetime(filtered_df['date'], dayfirst=True)


filtered_df.set_index('date', inplace=True)
```

```python
# Function to resample and aggregate weekly counts for each police force
def resample_weekly_counts(df, police_force):
    weekly_data = df[df['police_force'] == police_force].resample('W').size()
    return weekly_data

# Create a DataFrame to hold the weekly counts for each police force
weekly_counts = pd.DataFrame()

for police_force in selected_police_forces:
    weekly_counts[f'police_force_{police_force}'] = resample_weekly_counts(filtered_df, police_force)

print(weekly_counts.head())
```

```python
import statsmodels.api as sm

# Function to create and fit ARIMA model
def fit_arima_model(data):
    model = sm.tsa.ARIMA(data, order=(5, 1, 0))  # ARIMA(p, d, q) model
    model_fit = model.fit()
    return model_fit

# Fit ARIMA models for each police force
models = {}
for col in weekly_counts.columns:
    models[col] = fit_arima_model(weekly_counts[col])

# Forecast for the next 52 weeks (one year)
forecast_period = 52
forecasts = {}
for col in weekly_counts.columns:
    forecast_result = models[col].get_forecast(steps=forecast_period)
    forecasts[col] = forecast_result.predicted_mean  # Get the forecasted mean values

# Create DataFrame to hold forecasts
forecast_df = pd.DataFrame(forecasts, index=pd.date_range(start=weekly_counts.index[-1] + pd.Timedelta(weeks=1), periods=forecast_period, freq='W'))

print(forecast_df.head())
```

```python
# Plot historical and forecasted data
plt.figure(figsize=(15, 10))
for col in weekly_counts.columns:
    plt.plot(weekly_counts[col], label=f'Historical - {col}')
    plt.plot(forecast_df[col], label=f'Forecast - {col}', linestyle='--')

plt.title('Weekly Accident Counts and Forecasts')
plt.xlabel('Date')
plt.ylabel('Number of Accidents')
plt.legend()
plt.show()
```

```python
# Load the accident data from the database
accidents_query = "SELECT * FROM accident WHERE local_authority_ons_district = 'E06000010';"
accidents_df = pd.read_sql_query(accidents_query, connection)

# Filter data for the first three months of 2020
accidents_df['date'] = pd.to_datetime(accidents_df['date'], dayfirst=True)
first_three_months_df = accidents_df[(accidents_df['date'] >= '2020-01-01') & (accidents_df['date'] <= '2020-03-31')]

# Identify the three LSOAs with the highest number of accidents
top_lsoas = first_three_months_df['lsoa_of_accident_location'].value_counts().head(3).index.tolist()
print(f"Top 3 LSOAs with highest accidents: {top_lsoas}")
```

```python

# Filter data for the top LSOAs for the last six months (Jan to June 2020)
six_months_df = accidents_df[(accidents_df['date'] >= '2020-01-01') & (accidents_df['date'] <= '2020-06-30')]

# Function to create and fit ARIMA model
def fit_arima_model(data):
    model = sm.tsa.ARIMA(data, order=(5, 1, 0))  # ARIMA(p, d, q) model
    model_fit = model.fit()
    return model_fit

forecasts = {}

for lsoa in top_lsoas:
    lsoa_data = six_months_df[six_months_df['lsoa_of_accident_location'] == lsoa]
    
    # Aggregate to daily accident counts
    daily_counts = lsoa_data.resample('D', on='date').size()
    
    # Fit the ARIMA model
    model = fit_arima_model(daily_counts)
    
    # Forecast for the next month (31 days)
    forecast_period = 31
    forecast_result = model.get_forecast(steps=forecast_period)
    forecast = forecast_result.predicted_mean
    
    # Store the forecast in the dictionary
    forecasts[lsoa] = forecast

# Convert the forecast dictionary to a DataFrame
forecast_df = pd.DataFrame(forecasts)
print(forecast_df.head())
print(forecast_df.tail())


```

```python
# Plotting the forecasts
plt.figure(figsize=(15, 10))
for lsoa in forecasts:
    plt.plot(forecast_df[lsoa], label=f'Forecast - {lsoa}')

plt.title('Daily Accident Forecast for July 2020')
plt.xlabel('Date')
plt.ylabel('Number of Accidents')
plt.legend()
plt.show()
```
